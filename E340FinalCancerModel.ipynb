## Biological Background

Cancer is a genetic disease that is caused by changes to genes that control the way our cells function in how they grow and divide. Cancer cells differ from normal cells in that they are less specialized and do not mature into cells with a specific type or function and they also ignore signals that usually tell cells to stop dividing or when to undergo apoptosis, which is programed cell death. 
Tumors are often associated with cancer as a tumor is an abnormal mass of tissues that form when cells grow and divide more than they should and also do not die when they should which can cause large masses to form. Tumors can either be Benign, non-cancerous, or malignant, cancerous, and this application and model focuses on the second one. Tumors can also use the immune system to stay alive and grow. 


### Immune System 

The immune system is how they body naturally fights to protect the body as its natural defensive system. It is made up of a network of cells, molecules, tissues, and organs that all work together to help prevent and fight infection and diseases, including cancer. The type of cells that are included in the immune system include: B Cells, several types of T cells, Dendritic cells, and Macrophages. B cell are cells that release antibodies (proteins that bind to specific antigen markers that mark which cells are harmful) that defend against invading cells, tumor-reactive B cells bind to cancer cells and disrupt their activity as it also stimulates an immune response against the tumor cells. One type of T cell is the CD8+ killer T cells that destroy virus-infected cells. These cells can also directly target cancer cells to destroy them and CD8+ cells can also be called to help from CD4+ cell that send out signals to other immune cells to help direct their responses to destroy cells more efficiently. Dendritic cells are able to digest cancerous cells by presenting their proteins on surfaces where other immune cells can better recognize harmful cells they need to destroy. One molecule that is important in the immune system besides antibodies is Cytokines, messenger molecules that help immune cells coordinate and work together for the best immune response to the tumor.

Cells can die either immunogenically or non-immunogenically. Immunogenic cell death is apoptosis (cell death) resulting in a regulated activation of the immune response and immunogenic death of a cancer cell can induce an anti-tumor immune response through the activation of dendritic cells which then activate the specific T cell response needed to go after the tumor. Non-immunogenic cell death is apoptosis that helps with homeostasis of the body, is cell death that goes unnoticed by the immune system and is just part of the cells life. While it is important to have that type of cell death, the recent studies in immunogenic cell death is a huge part of immunotherapy treatment. 



### Radiation

There are many treatments of cancer and one of the most common is Radiation, a treatment that used high doses of radiation to kill and shrink tumors. Radiation is energy that is carries by waves or a stream of particles that damages the DNA in cells. Cancer cells whose DNA gets too damaged cannot be repaired stop dividing and die. Radiation first kills cells that are actively dividing in the cell cycle and cancer cells are expected to divide quickly and out of control and this is helpful for killing them. However, Radiation also harms normal dividing tissues which can cause unwanted damages and side effect so there much be a balance between destroying cancer cells and not too many of normal cells.  The radiation can induce cell stress and an immunogenic cell death that can then active antigen-presenting cells (APCs), once active the APC can then travel to active CD8+ T cells that can fight against the tumor that have cells with those antigens present. These cells are now distributed throughout the body and can find other tumor sites that arent being treated directly with radiation but still have those antigens and can be killed, this is also known as the abscopal effect. Radiation treatment can take days or weeks. Cancer cells then continue to die for weeks to months after and usually only treats cancer not cure it. 


### Immunotherapy 

Immunotherapy is a cancer treatment that uses the power of ones own immune system to prevent, control and eliminate cancer. Its a biological therapy, as it uses substances made from living organisms, that can help the immune system learn, recognize and attack specific cancer cells, boost immune cells to help them eliminate cancer, and provide the body with additional components to enhance the immune response. There are a variety of forms that immunotherapy comes in such as targeted antibodies, cancer vaccines, adoptive cell transfer, tumor-infecting viruses, checkpoint inhibitors, cytokines, and adjuvants. When immunotherapy works it has immune cells produce cytokines (protein molecules that act on the other cells) in large amount into the body which stimulates the immune system and kicks it into high gear. Immunotherapy does not work for everyone it is often used with another form of cancer treatment as it does in this model with radiation. 

## Mathematical Representation 

The main goal of the mathematical model is to simulate the radiation response and the induced immune response as well as the response from an immunotherapy dose. In the paper they use a logistic equation model of the tumor with a carrying capacity K, this is because a tumor cannot grow forever. It would not be able to be an exponential growth model because of the resources in the body. While at the beginning of the model it follows an exponential growth there is an inflection point where the competition of resources comes into play and a logistic model follows that trend. This inflection point can be influenced both by the carrying capacity, k, and by the growth rate, r.

### Important Parameters from Equations

This is a list of the important parameters from the paper and equations above an also the equations implimented in each simulations. The variable names are listed, the units, and the starting values. Many of the values are able to be adjusted throughout the model and on each simulation. 

- r : viable Cancer cells volume doubling time (1/day), 0.195
- K : Tumor carrying Capacity (mm^3), 1423
- a : CTL's killing rate (mm^3/(cell x day), 0.0177
- d : clearence rate of dying cells (1/day), 0.264
- SFd : Fraction of cancer cells that survie radiation, 0.2
- l : decay rate of effector cells (1/day), 0.3
- w2 : change in baseline T cell recruitment due to immnogenic cell death, 15.37
- lambda1 : rate of how long immune system responses last, (1/day), 20
- Estar : equilibrium stregnth of immune system, 1
- strength: strength of the immunotherapy drug, 10

## Simulating NO immune system with 2 tumors growing Indpendently 

Goal: for this simulation it will allow the user to change the growth rate, r, and the carrying capacity, k, individually for both of the tumors independently and to see what happens when there is no immune systme or cancer treatment to stop the cancer cells, C, from growing. 


```python
# Simulation with two tumors growing independently without an immune system

base_model = '''
    //Equations 
    
    //Primary tumor 
    
    E1: -> C1; r1*C1*(1- (V1/K1)); // tumor growth rate
    E2: D1 ->; d*D1 ; //clearance of dead cells
    
    //Secondary Tumor
    
    
    E3: -> C2; r2*C2*(1- (V2/K2)); // tumor growth rate
    E4: D2 ->; d*D2 ; //clearance of dead cells
    
    //Primary tumor
    V1 := C1 + D1 //Total Volume Primary Tumor
    
    //Secondary Tumor
    V2:= C2 + D2 //Total Volume Secondary Tumor
    
    
    //Parameters
    r1 = 0.195 // cancer cells growth rate Primary Tumor
    K1 = 1423 // tumor carrying capacity of Primary tumor 
    d = 0.264 //clearance rate of dying cells 
    r2 = 0.195 // cancer cells growth rate Secondary Tumor
    K2 = 725 // tumor carrying capacity of Secondary tumor


    //Initial Conditions
    C1 = 1 // initial cancer cells of primary tumor 
    D1 = 0 // initial non-immunogenic dying cancer cells
    C2 = 1 // initial cancer cells of primary tumor 
    D2 = 0 // initial non-immunogenic dying cancer cells    

'''

m = te.loada(base_model)
s = m.simulate(0,100,1000)
m.plot(title="Cell population of all cell types (r = .195, K1 = 1423, K2 = 725)", xtitle="time (days)", ytitle="Cell Population",linewidth=2)



# Slider changing r, growth rate for each tumor 

from ipywidgets import *
def replot(rT1,rT2):
    m = te.loada(base_model)
    m.r1 = rT1
    m.r2 = rT2
    s = m.simulate(0,100,1000,['time','C1','D1','C2','D2'])
    m.plot(title="Growth Rate (r) Adjustment for Both Tumors", xtitle="time (days)", ytitle="Cell Population",ylim=(0,1500))
    print('Growth rate of Primary tumor =',m.r1)
    print('Growth rate of Secondary Tumor =', m.r2)

interact(replot,rT1=FloatSlider(min=0, max=0.5, step=0.01, value=0, continuous_update = False), rT2=FloatSlider(min=0, max=0.5, step=0.01, value=0,continuous_update = False))

# Slider changing tumor Carrying capacity for each tumor 

def replot(KT1,KT2):
    m = te.loada(base_model)
    m.K1 = KT1
    m.K2 = KT2
    s = m.simulate(0,100,1000,['time','C1','D1','C2','D2'])
    m.plot(title="Carying Capacity of Tumor Sites", xtitle="time (days)", ytitle="Cell Population",ylim=(0,1500))
    print('Carrying Capacity of Primary Tumor =' ,m.K1)
    print('Carrying Capacity of Secondary Tumor =' ,m.K2)

interact(replot,KT1=FloatSlider(min=1, max=1423, step=1, value=0, continuous_update = False), KT2=FloatSlider(min=1, max=1423, step=1, value=0,continuous_update = False))
```

## What did you see?

Both tumors grow to their respective max number of cancer cells and there is no cell death without the immune system or cancer treatment in the simulation. 

### What happened when you changed the Growth Rate, r?

When you increase the growth rate, r, the time it takes for each respective tumor to reach its full size with max amount of tumor cells decreases and it grows faster. This is the oppostie case when you decrease the rate. When decreasing r, the tumor now takes longer to reach its max carrying capacity. This relationship is the same for both tumors.

### What happened when you changed the Carrying Capacity, K?

When changing the carrying capacity, K, for both tumors you can see that the cell population of cancer cells will increase when you increase max carrying capacity resulting in an over all bigger tumor and when you decrease the carrying capacity you see that the tumor gets smaller as you lower the max number of cancer cells the tumor can hold. This is the same regardless of the tumor site.  



## Simulating effect of Radiation on Primary tumor with NO immune response

Goal: This simulation is to show what happens when the primary tumor gets radiation and there is no immune system to response to the radiation and how the secondary tumor is effected.

```python
# Model with only radiation on primary tumor 


rad_model = '''
    //Equations (first tumor)
    
    //First tumor
    
    E1: -> C1; r1*C1*(1- (V1/K1)); // tumor growth rate
    E2: D1 ->; d*D1 ; //clearance of dead cells 
    E3: I1 ->; d*I1; //death by radiation
    
    //Secondary tumor
    
    E4: -> C2; r2*C2*(1- (V2/K2)); // tumor growth rate
    E5: D2 ->; d*D2 ; //clearance of dead cells 
    
    
    //First tumor
    V1 := C1 + D1 + I1
    Cdeath1 := C1*SFd // dying from radiation
    
    at(time > t1) : C1 = Cdeath1, I1 = I1 + Cdeath1 ; //once add radiation then I1 present
 
    //Secondary tumor
    
    V2 := C2 + D2 // total volume of Secondary tumor
    
    
    //Parameters
    r1 = 0.195 // cancer cells growth rate Primary Tumor
    K1 = 1423 // tumor carrying capacity
    a = 0.0177 //CTL's killing rate
    d = 0.264 //clearance rate of dying cells 
    SFd = 0.2 //radition kills % - mention 1 - .8 is how got .2
    t1 = 60 // time of radiation treatment 
    r2 = 0.195 // cancer cells growth rate Secondary Tumor
    K2 = 725 // tumor carrying capacity of Secondary tumor

    nDays = 100 //number of days 

    //Initial Conditions
    C1 = 5 // initial cancer cells of primary tumor 
    D1 = 0 // initial non-immunogenic dying cancer cells
    I1 = 0 // initial dying cells from radiaition
    
    
    C2 = 1 // initial cancer cells of 2nd tumor 
    D2 = 0 // initial non-immunogenic dying cancer cells

'''

m = te.loada(rad_model)
s = m.simulate(0,100,1000)
m.plot(title="Cell population when Radiation done at t1", xtitle="time (days)", ytitle="Cell Population",linewidth=2)


# Slider changing duration of simulation
def replot(xDays):
    m = te.loada(rad_model)
    m.nDays = xDays
    s = m.simulate(0,xDays,1000,['time','C1','D1','I1','C2','D2'])
    m.plot(title="Cell population when Radiation done at t1", xtitle="time (days)", ytitle="Cell Population",ylim=(0,1500))
    print('Day simulation ran =', m.nDays)

interact(replot,xDays=FloatSlider(min=100, max=739, step=10, value=0, continuous_update = False))
```

## What did you see? 

### What happened to the Tumor?  How long did it take for the tumor to get back to the original size?

The primary tumor shrunk at the time of radiation, 60 days, and then it rapidly grew back to its original size and stayed at it's max number of cancer cells. While the first tumor was shrunk momentarily you can notice that the second tumor stayed unaffected without the immune system. 

### What happened when you changed How long the simulation was ran?

In this case making the simulation run longer showed how that once radiation was done and the tumor grew quickly back to its original size there was no change after and it stayed consistently at max carrying capacity and the secondary tumor never had a change regardless of how long you ran the simulation. 

## Simulating effect of immune system WITHOUT immunotherapy

Goal: This simulation is to show the immune system effect without immunotherapy and without radiation to see how the body naturally defends itself to the cancer cells. The user will be able to adjust both the equilibrium point of the immune system and the the efficiency  of the immune system. 

```python
# Model with immune system but NO immunotherapy (or radiation)

ImmuneSystem_model = '''
    //Equations (first tumor)
    
    //First tumor
    
    E1a: -> C1; r1*C1*(1- (V1/K1)); // tumor growth rate with resource limitation
    E2a: C1 -> D1; a*C1*E1 ; //tumor clearance by immune cells
    E3: D1 ->; d*D1 ; //clearance of dead cells
    E4: -> E1; lambda1*Estar - lambda1*E1 + w2*(I1)*(V1/(V1+V2)) ; //cytotoxic T cells from immmune system effects
    E5: E1 ->; l*E1*V1; // decay

  
    
    //Secondary tumor
    
    E7: -> C2; r2*C2 * (1-V2/K2); // tumor growth rate
    E8: C2 -> D2; a*C2*E2 ; //tumor clearance by immune cells 
    E9: D2 ->; d*D2 ; //clearance of dead cells
    E10: -> E2; lambda1*Estar - lambda1*E2 + w2*(I1)*(V2/(V1+V2)) ; //cytotoxic T cells from immmune system effects
    E11: E2 ->; l*E2*V2; // decay
    
    
    //First tumor
    V1 := C1 + D1 ;  // total volume of cells 
 
    //Secondary tumor
    
    V2 := C2 + D2 ;  // total volume of cells secondary tumor

    
    //Parameters
    r1 = 0.195 // cancer cells growth rate Primary Tumor
    K1 = 1423 // tumor carrying capacity
    a = 0.0177 //CTL's killing rate
    d = 0.264 //clearance rate of dying cells 
    l = 0.03 // decay rate of effector cells 
    w2 = 15.37 //change in baseline T cell recruitment due to immnogenic cell death 
    t1 = 60 // time of radiation treatment 
    lambda1 = 20 // how long a immune response reaction lasts
    Estar = 1 // sets equiliburim point for immune response
    r2 = 0.195 // cancer cells growth rate Secondary Tumor
    K2 = 725 // tumor carrying capacity of Secondary tumor
    
    
    nDays = 100 //number of days 

    //Initial Conditions
    C1 = 1 // initial cancer cells
    D1 = 0 // initial non-immunogenic dying cancer cells
    E1 = 0 // initial amount of tumor specific cytotoxic T cells 
    I1 = 0 // initial immunogenic dying cells
    
    
    C2 = 1 // initial cancer cells
    D2 = 0 // initial non-immunogenic dying cancer cells
    E2 = 0 // initial amount of tumor specific cytotoxic T cells 


'''

m = te.loada(ImmuneSystem_model)
s = m.simulate(0,150,1000)
m.plot(title="Cell population with Immune System (E* at 1)", xtitle="time (days)", ytitle="Cell Population",linewidth=2)

# Slider changing E* (Estar) and time duration of simulation
def replot(eStar, xDays):
    m = te.loada(ImmuneSystem_model)
    m.a = 1.25*m.a
    m.nDays = xDays
    m.Estar = eStar
    s = m.simulate(0,xDays,1000,['time','C1','D1','E1','I1','C2','D2','E2'])
    m.plot(title="Cell population when Adjusting E*", xtitle="time (days)", ytitle="Cell Population",ylim=(0,1400))
    print('E* Equilibrium Point =', m.Estar)
    print('Day simulation ran =', m.nDays)

interact(replot,eStar=FloatSlider(min=0, max= 10, step=.5, value=0, continuous_update = False),xDays=FloatSlider(min=100, max=739, step=10, value=0, continuous_update = False))


# Slider changing E* (Estar), a, and time duration of simulation
def replot(eStar, efficiency, xDays):
    m = te.loada(ImmuneSystem_model)
    m.a = efficiency*m.a
    m.nDays = xDays
    m.Estar = eStar
    s = m.simulate(0,xDays,1000,['time','C1','D1','E1','I1','C2','D2','E2'])
    m.plot(title="Cell population when Adjusting E* and Immune Efficiency, a", xtitle="time (days)", ytitle="Cell Population",ylim=(0,1400))
    print('E* Equilibrium Point =', m.Estar)
    print('Immune Efficiency =', m.a)
    print('Day simulation ran =', m.nDays)
    

interact(replot,eStar=FloatSlider(min=0, max= 10, step=.5, value=0, continuous_update = False), efficiency =FloatSlider(min=1, max= 10, step=.5, value=0, continuous_update = False),xDays=FloatSlider(min=100, max=739, step=10, value=0, continuous_update = False))


```

## What did you see?

When adding in the immune system allowed for some amount of death cells to come in and help shrink the number of cancer cells and tumor sizes.

### What happened to BOTH tumors?

When you increase the immune system Equilibrium point (E*) both tumors take longer to grow and increase in number of cancer cells. When decreasing the equilibrium point the tumor grows much faster.

### Did the Immune system reduce the size of the tumors, if so how much?

The immune system would decrease the overall tumor sizes as you increase the equilibrium point as well as take longer to grow. And just like when you decreased E* and the tumors grew faster there were also more cancer cells and two bigger tumors sites. The amount both tumor sizes decreased was in proportional to how much you increased the equilibrium point so the larger the Estar the more the tumors shrunk and the less cancer cells there were. 

### What happened when you changed how long the simulation was ran?

In this simulation when you increased the equilibrium point and there for made the tumors take longer to grow you would have to run the simulation for longer as the immune system now was fighting the cancer cells for longer and it took more to overcome the immune system.

### What was the relationship between immune efficiency (a) and equilibrium point (E*)

Both the immune efficiency (a) and equilibrium point (Estar) help fight off the tumor and decrease cancer cells, so when you increase one you will need less of the other. The more the immune efficiency was increased the lower of an equilibrium point was needed to help eliminate the tumors. 

## Simulating of Immunotherapy with NO Radiation 

Goal: This simulation is to show how immunotherapy and NO Radiation works on the cancer cells. USer will be able to adjust both the immune efficiency and the strength of the immunotherapy drug

```python
# Model immunotherapy with NO radiaiton 

immunotherapy_model = '''
    //Equations (first tumor)
    
    //First tumor
    
    E1a: -> C1; r1*C1*(1- (V1/K1)); // tumor growth rate
    E2a: C1 -> D1; a*C1*E1 ; //tumor clearance by immune cells and recources
    E3: D1 ->; d*D1 ; //clearance of dead cells
    E4: -> E1; lambda1*Estar*(1+ U1) - lambda1*E1; // cytotoxic T cells from immmuotherapy effects
    E5: E1 ->; l*E1*V1; // decay
 
    
    //Secondary tumor
    
    E7: -> C2; r2*C2 * (1-V2/K2); // tumor growth rate
    E8: C2 -> D2; a*C2*E2 ; //tumor clearance by immune cells 
    E9: D2 ->; d*D2 ; //clearance of dead cells
    E10: -> E2; lambda1*Estar*(1 + U1) - lambda1*E2  ; //cytotoxic T cells from immmuotherapy effects
    E11: E2 ->; l*E2*V2; // decay

    E13: U1 ->; (1/LifeTime) * U1 ; //immuntotherapy component
    
    
    //First tumor
    V1 := C1 + D1 + E1 ;  // total volume of cells 

    //Secondary tumor
    
    V2 := C2 + D2 + E2 ;
        

    at(time > t2) : U1 = U1 + strength ; // immunotherapy 
    
    //Parameters
    r1 = 0.195 // cancer cells growth rate Secondary Tumor
    K1 = 1423 // tumor carrying capacity
    a = 0.0177 //CTL's killing rate
    d = 0.264 //clearance rate of dying cells 
    w = 0.135 //Baseline T cell recruitment rate
    w2 = 15.37 //change in baseline T cell recruitment due to immnogenic cell death 
    l = 0.03 // decay rate of effector cells 
    SFd = 0.2 //radition kills % (change) - mentions 1 - .8 is how got .2 
    t2 = 60 // time of immunotherapy 
    lambda1 = 60 // how long a immune response reaction lasts
    Estar = 1 // sets equiliburim point for immune response
    LifeTime = 10 // Life Time of days that the immune response happens 
    r2 = 0.195 // cancer cells growth rate Secondary Tumor
    K2 = 725 // tumor carrying capacity of Secondary tumor
    strength = 10 // strength of immunotherapy

    nDays = 100 //number of days 

    //Initial Conditions
    C1 = 1 // initial cancer cells
    D1 = 0 // initial non-immunogenic dying cancer cells
    E1 = 0 // initial amount of tumor specific cytotoxic T cells 

    
    
    C2 = 1 // initial cancer cells
    D2 = 0 // initial non-immunogenic dying cancer cells
    E2 = 0 // initial amount of tumor specific cytotoxic T cells 


'''

m = te.loada(immunotherapy_model)
m.a = m.a* 3.0
s = m.simulate(0,200,1000)
m.plot(title="Cell population with Immunotherapy and NO Radiation", xtitle="time (days)", ytitle="Cell Population",linewidth=2)

#add slider of strength

# Slider changing a and time duration of simulation
def replot(efficiency, xDays):
    m = te.loada(immunotherapy_model)
    m.a = efficiency*m.a
    s = m.simulate(0,xDays,1000,['time','C1','D1','E1','C2','D2','E2'])
    m.plot(title="Cell population when Adjusting Immune Efficiency", xtitle="time (days)", ytitle="Cell Population")
    print('Immune Efficiency =', m.a)
    print('Day simulation ran =', m.nDays)
    

interact(replot, efficiency =FloatSlider(min=1, max= 10, step=.5, value=0, continuous_update = False),xDays=FloatSlider(min=100, max=739, step=10, value=0, continuous_update = False))


# Slider changing a and strength and time duration of simulation
def replot(efficiency, streng, xDays):
    m = te.loada(immunotherapy_model)
    m.a = efficiency*m.a
    m.strength = streng * m.strength
    m.nDays = xDays
    s = m.simulate(0,xDays,1000,['time','C1','D1','E1','C2','D2','E2'])
    m.plot(title="Cell population when Immune Efficiency and Immunotherapy Strength", xtitle="time (days)", ytitle="Cell Population",ylim=(0,1400))
    print('Immune Efficiency =', m.a)
    print('Strength of Immunotherapy =', m.Estar)
    print('Day simulation ran =', m.nDays)
    

interact(replot,streng=FloatSlider(min=0, max= 10, step=.5, value=0, continuous_update = False), efficiency =FloatSlider(min=1, max= 10, step=.5, value=0, continuous_update = False),xDays=FloatSlider(min=100, max=739, step=10, value=0, continuous_update = False))




```

## Simulating Radiation on Primary Tumor to See Effect on Secondary Tumor

User will be able to change both a, the immune system efficiency, and w2, how strongly the immune system reacts, individually and simultaniously to see their relationship. 

```python
# Model of effect of radiation on the primary exciting the immune response


RadEffect_model = '''
    //Equations (first tumor)
    
    //First tumor
    
    E1a: -> C1; r1*C1*(1- (V1/K1)); // tumor growth rate
    E2a: C1 -> D1;  a*C1*E1 ; //tumor clearance by immune cells and recources
    E3: D1 ->; d*D1 ; //clearance of dead cells
    E4: -> E1; lambda1*Estar*(1+U1) - lambda1*E1 + w2*(I1)*(V1/(V1+V2)) ; //cytotoxic T cells from immmune system effects
    E5: E1 ->; l*E1*V1; // decay
    E6: I1 ->; d*I1; //radiation
  
    
    //Secondary tumor
    
    E7: -> C2; r2*C2 * (1-V2/K2); // tumor growth rate
    E8: C2 -> D2;  a*C2*E2 ; //tumor clearance by immune cells 
    E9: D2 ->; d*D2 ; //clearance of dead cells
    E10: -> E2; lambda1*Estar*(1+U1) - lambda1*E2 + w2*(I1)*(V2/(V1+V2)) ; //cytotoxic T cells from immmune system effects
    E11: E2 ->; l*E2*V2; // decay

    
    
    //First tumor
    V1 := C1 + D1 + I1 ;  // total volume of cells 
    Cdeath1 := C1*SFd ;// dying immuogen.
    
    at(time > t1) : C1 = Cdeath1, I1 = I1 + Cdeath1 ;
 
    //Secondary tumor
    
    V2 := C2 + D2 ;

    
    //Parameters
    r1 = 0.195 // cancer cells growth rate Primary Tumor
    K1 = 1423 // tumor carrying capacity
    a = 0.0177 //CTL's killing rate
    d = 0.264 //clearance rate of dying cells 
    l = 0.03 // decay rate of effector cells 
    SFd = 0.2 //radition kills % (change) - mentions 1 - .8 is how got .2
    w2 = 15.37 //change in baseline T cell recruitment due to immnogenic cell death 
    t1 = 60 // time of radiation treatment 
    lambda1 = 60 // how long a immune response reaction lasts
    Estar = 1 // sets equiliburim point for immune response
    r2 = 0.195 // cancer cells growth rate Secondary Tumor
    K2 = 725 // tumor carrying capacity of Secondary tumor
    
    
    nDays = 100 //number of days 

    //Initial Conditions
    C1 = 1 // initial cancer cells
    D1 = 0 // initial non-immunogenic dying cancer cells
    E1 = 0 // initial amount of tumor specific cytotoxic T cells 
    I1 = 0 // initial immunogenic dying cells
    
    
    C2 = 1 // initial cancer cells
    D2 = 0 // initial non-immunogenic dying cancer cells
    E2 = 0 // initial amount of tumor specific cytotoxic T cells 
    U1 = 0 // no immunotherapy 

'''

m = te.loada(RadEffect_model)
s = m.simulate(0,150,1000)
m.plot(title="Cell population with Immune System and Radiation", xtitle="time (days)", ytitle="variable",linewidth=2)


m = te.loada(RadEffect_model)
s = m.simulate(0,150,1000, ['time', 'C2','E2'])
m.plot(title="Cell population with Immune System and Radiation)", xtitle="time (days)", ytitle="variable",linewidth=2)



# slider of a 
def replot(efficiency, xDays):
    m = te.loada(RadEffect_model)
    m.a = efficiency*m.a
    m.nDays = xDays
    s = m.simulate(0,xDays,1000,['time','C2','E2'])
    m.plot(title="Cell population when Radiation on Primary Tumor Adjusting Immune Efficiency", xtitle="time (days)", ytitle="Cell Population",ylim=(0,700))
    print('Immune Efficiency =', m.a)
    print('Day simulation ran =', m.nDays)
    
interact(replot,efficiency =FloatSlider(min=1, max= 10, step=.5, value=0, continuous_update = False), xDays=FloatSlider(min=100, max=739, step=10, value=0, continuous_update = False))

# slider of w2 
def replot(react, xDays):
    m = te.loada(RadEffect_model)
    m.w2 = react*m.w2
    m.nDays = xDays
    s = m.simulate(0,xDays,1000,['time','C2','E2'])
    m.plot(title="Cell population when Radiation on Primary Tumor Adjusting Immune Reaction", xtitle="time (days)", ytitle="Cell Population",ylim=(0,700))
    print('Strength of Immune System Reaction =', m.w2)
    print('Day simulation ran =', m.nDays)
    
interact(replot,react =FloatSlider(min=1, max= 10, step=.5, value=0, continuous_update = False), xDays=FloatSlider(min=100, max=739, step=10, value=0, continuous_update = False))



# combination slider of w2 and a
def replot(efficiency,react,xDays):
    m = te.loada(RadEffect_model)
    m.a = efficiency*m.a
    m.w2 = react*m.w2
    m.nDays = xDays
    s = m.simulate(0,xDays,1000,['time','C2','E2'])
    m.plot(title="Cell population when Radiation on Primary Tumor Adjusting Immune Reaction and Immune Efficiency", xtitle="time (days)", ytitle="Cell Population",ylim=(0,700))
    print('Immune Efficiency =', m.a)
    print('Strength of Immune System Reaction =', m.w2)
    print('Day simulation ran =', m.nDays)
    
interact(replot,efficiency =FloatSlider(min=1, max= 10, step=.5, value=0, continuous_update = False), react =FloatSlider(min=1, max= 10, step=.5, value=0, continuous_update = False), xDays=FloatSlider(min=100, max=739, step=10, value=0, continuous_update = False))

```

## Simulating Radiation on Primary Tumor with Immunotherapy

They will be able to adjust both the amount of Radiation and the Immunotherapy Effect to see how the seconary tumor cancer cells are effected. 

```python
# Model of Immunotherapy AND radiation 

both_model = '''
    //Equations (first tumor)
    
    //First tumor
    
    E1a: -> C1; r1*C1*(1- (V1/K1)); // tumor growth rate
    E2a: C1 -> D1; a*C1*E1 ; //tumor clearance by immune cells
    E3: D1 ->; d*D1 ; //clearance of dead cells
    E4: -> E1; lambda1*Estar*(1+U1) - lambda1*E1 + (1 + U1)*w2*(I1)*(V1/(V1+V2))  ; // cytotoxic T cells
    E5: E1 ->; l*E1*V1; // decay
    E6: I1 ->; d*I1; //radiation
  
    
    //Secondary tumor
    
    E7: -> C2; r2*C2 * (1-V2/K2); // tumor growth rate
    E8: C2 -> D2; a*C2*E2 ; //tumor clearance by immune cells 
    E9: D2 ->; d*D2 ; //clearance of dead cells
    E10: -> E2; lambda1*Estar * (1 +U1) - lambda1*E2 + (1 + U1)*w2*(I1)*(V2/(V1+V2)) ; //cytotoxic T cells 
    E11: E2 ->; l*E2*V2; // decay
    E12: I2 ->; d*I2; //radiation
    
    E13: U1 ->; (1/LifeTime) * U1 ; //immunotherapy component
    
    
    //First tumor
    V1 := C1 + D1 + I1 ;  // total volume of cells 
    Cdeath1 := C1*SFd ;// dying immuogen.
    
    
    //Radiation 

    at(time > t1) : C1 = Cdeath1, I1 = I1 + Cdeath1 ;
 
    //Secondary tumor
    
    V2 := C2 + D2 ; // total volume of cells in secondary tumor 
    
    //Immunotherapy 
    
    at(time > t2) : U1 = U1 + strength ; // immunotherapy 
    
    //Parameters
    r1 = 0.195 // viable cancer cells volume doubling time
    K1 = 1423 // tumor carrying capacity
    a = 0.0177 //CTL's killing rate
    d = 0.264 //clearance rate of dying cells 
    w = 0.135 //Baseline T cell recruitment rate
    w2 = 15.37 //change in baseline T cell recruitment due to immnogenic cell death 
    l = 0.03 // decay rate of effector cells 
    SFd = 0.2 //radition kills % - mentions 1 - .8 is how got .2
    t1 = 60 // time of radiation treatment 
    t2 = 59 // time of immunotherapy 
    lambda1 = 60 // how long a immune response reaction lasts
    Estar = 2 // sets equilliburim point for immune response
    LifeTime = 10 // Life Time of days that the immune response happens 
    r2 = 0.195 // cancer cells growth rate Secondary Tumor
    K2 = 725 // tumor carrying capacity of Secondary tumor
    strength = 10 // strength of Immunotherapy


    nDays = 100 //number of days 

    //Initial Conditions
    C1 = 1 // initial cancer cells
    D1 = 0 // initial non-immunogenic dying cancer cells
    E1 = 0 // initial amount of tumor specific cytotoxic T cells 
    I1 = 0 // initial immunogenic dying cells
    
    
    C2 = 1 // initial cancer cells
    D2 = 0 // initial non-immunogenic dying cancer cells
    E2 = 0 // initial amount of tumor specific cytotoxic T cells 


'''

m = te.loada(both_model)
s = m.simulate(0,200,1000)
m.plot(title="Cell population with BOTH radiation and Immunotherapy", xtitle="time (days)", ytitle="Cell Population",linewidth=2)

m = te.loada(both_model)
#m.w2 = 5.0 * m.w2
#m.a = 3.0 * m.a
#m.strength = m.strength * 1.0
s = m.simulate(0,200,1000, ['time', 'C2','E2'])
m.plot(title="Cell population Secondary Tumor with BOTH radiation and Immunotherapy", xtitle="time (days)", ytitle="variable",linewidth=2)
print('Immunotherapy Effect =', m.lambda1)
print('Radiaiton = ', m.SFd)

#Slider that changes lambda1 (immunotherapy) and SFd (Radiaiton amount)
def replot(Lambda,Radiation,xDays):
    m = te.loada(both_model)
    m.lambda1 = Lambda
    m.SFd = Radiation
    m.nDays = xDays
    s = m.simulate(0,xDays,1000,['time','C1','E1','C2','E2'])
    m.plot(title="Cell Populations of Secondary Tumor adjusting amount of Radiation and Immunotherapy", xtitle="time (days)", ytitle="Cell Population",ylim=(0,1400))
    print('Immunotherapy Effect = ',m.lambda1)
    print('Radiaiton = ', m.SFd)
    print('Day simulation ran =', m.nDays)

interact(replot,Lambda=FloatSlider(min=5, max=100, step=1, value=0, continuous_update = False), Radiation=FloatSlider(min=0, max=0.7, step=0.01, value=0,continuous_update = False),xDays=FloatSlider(min=100, max=739, step=10, value=0, continuous_update = False))
```
